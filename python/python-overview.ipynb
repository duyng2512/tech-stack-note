**Memory management of python**:

- Python is an object-oriented programming language. Everything is in Python treated as an object, including variable, function, list, tuple, dictionary, set, etc. Every object belongs to its class. An object is a real-life entity. An object is the collection of various data and functions that operate on those data.

**Garbage Collectors:**
- Python use a reference counter to object to determine it should be collected or not, if there is zero reference to object it should be collected.

There are two aspects to memory management and garbage collection in CPython:

- Reference counting: 
    - At a very basic level, a Python objects reference count is incremented whenever the object is referenced, and its decremented when an object is dereferenced. If an objects reference count is 0, the memory for the object is deallocated. Some people claim reference counting is a poor mans garbage collector. It does have some downsides, including an inability to detect cyclic references as discussed below. However, reference counting is nice because you can immediately remove an object when it has no references.
- Generational garbage collection:
    - Object in python start its life cycle in the first generation of the garbage collector.  If Python executes a garbage collection process on a generation and an object survives, it moves up into a second, older generation. The Python garbage collector has three generations in total, and an object moves into an older generation whenever it survives a garbage collection process on its current generation.
    - We can configure the threshold of each generatation, the larger the number the less frequency of collecting proccess.
